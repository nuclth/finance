Functions to get the filings for a given day from the Edgar daily index

```python
# import our libraries
import os
import requests
import pandas as pd
import datetime
from bs4 import BeautifulSoup

def make_url (base_url, comp):
    
    url = base_url
    
    for r in comp:
        url = '{}/{}'.format(url, r)
        
    return url

base_url = r'https://www.sec.gov/Archives/edgar/data'
components = ['886982', '000156459019011378', '0001564590-19-011378-index-headers.html']
make_url(base_url, components)
```

```python
base_url = r'https://www.sec.gov/Archives/edgar/daily-index'

year_url = make_url(base_url, ['2019', 'index.json'])
content = requests.get(year_url)
decoded_content = content.json()
```

```python
# loop through directory
for item in decoded_content['directory']['item']:

    # get name of folder
    print('-'*100)
    print('Pulling url for quarter {}'.format(item['name']))
    
    #create the qtr url 
    qtr_url = make_url(base_url, ['2019', item['name'], 'index.json'])
    
    # request file url and decode
    file_content = requests.get(qtr_url)
    decoded_file = file_content.json()
    
    print('_'*100)
    print('\tPulling Files')
    
    for file in decoded_file['directory']['item']:
        
        file_url = make_url(base_url, ['2019', item['name'], file['name']])
        print(file_url)
```

```python
# define a master file url
file_url = r'https://www.sec.gov/Archives/edgar/daily-index/2019/QTR2/master.20190401.idx'

# request that file
content = requests.get(file_url).content

# write to text
with open('master_20190401.txt', 'wb') as f:
    f.write(content)
```

```python
with open('master_20190401.txt', 'rb') as f:
    byte_data = f.read()

# decode the byte data
data = byte_data.decode('utf-8').split('  ')

for index, entry in enumerate(data):
    
    if 'ftp://ftp.sec.gov/edgar/' in entry:
        start_index = index
        
data_format = data[start_index + 1:]  # why the plus 1? Error in video?, seems like he's skipping things

master_data = []

for index, entry in enumerate(data_format):
    
    if index == 0:
        clean_entry_data = entry.replace('\n', '|').split('|')
        clean_entry_data = clean_entry_data[8:]
    else:
        clean_entry_data = entry.replace('\n', '|').split('|')
    
    for index, row in enumerate(clean_entry_data):
        
        # when you find the text file url
        if '.txt' in row:
            mini_list = clean_entry_data[(index-4):(index + 1)]
            
            if len(mini_list) != 0:
                mini_list[4] = 'https://www.sec.gov/Archives/' + mini_list[4]
                master_data.append(mini_list)
                
master_data[:3]
```

```python
master_dict = []

# loop through master data set
for index, document in enumerate(master_data):
    
    # create a dictionary
    document_dict = {}
    document_dict['cik_number'] = document[0]
    document_dict['company_name'] = document[1]
    document_dict['form_id'] = document[2]
    document_dict['date'] = document[3]
    document_dict['file_url'] = document[4]

    master_dict.append(document_dict)
```
